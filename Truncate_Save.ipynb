```python
import os
import pandas as pd

path_result='C:\\Users\\yunfa\\180b\\results\\'
path_data="C:\\Users\\yunfa\\180b\\data"
path_code="C:\\Users\\yunfa\\180b"
path_temp="C:\\Users\\yunfa\\180b\\temp"

schools = ["notredame","uofm","columbia","UCSD","berkeley","Harvard","ucla"]
s=schools[1]#use this index to control which network we are going to run.
t="2019"


# Total number of rows for each school and year as per the table
total_rows = {
    ('ucla', '2019'): 16894,
    ('ucla', '2020'): 18714,
    ('UCSD', '2019'): 17502,
    ('UCSD', '2020'): 17346,
    ('berkeley', '2019'): 16673,
    ('berkeley', '2020'): 18194,
    ("uofm", '2019'): 12524,
    ("uofm", '2020'): 14879,
    ('Harvard', '2019'): 2893,
    ('Harvard', '2020'): 3554,
    ('columbia', '2019'): 5031,
    ('columbia', '2020'): 9337,
    ('dartmouth', '2019'): 1274,
    ('dartmouth', '2020'): 1226,
    ('notredame', '2019'): 1579,
    ('notredame', '2020'): 7950,
}


# List of schools
schools = ["notredame", "uofm", "columbia", "UCSD", "berkeley", "Harvard", "ucla", "dartmouth"]

# Function to truncate and save the files
def truncate_and_save(school, year):
    # Change directory to the data path
    os.chdir(path_data)
    
    # File name pattern
    name = "label_emt3_" + school + year + "_cm.csv"
    
    # Read the CSV file
    truelabels = pd.read_csv(name, skip_blank_lines=True, header=None)
    
    # Truncate the DataFrame to the first "Total #" of rows
    total_number = total_rows[(school, year)]
    truelabels_truncated = truelabels.iloc[:total_number]
    
    # Save the truncated DataFrame back to the CSV file
    truelabels_truncated.to_csv(name, index=False, header=None)

# Loop over schools and years and process the files
for school in schools:
    for year in ['2019', '2020']:
        truncate_and_save(school, year)
```
